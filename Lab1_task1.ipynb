```python
import numpy as np

# Create a 4x4 array from exponential distribution with scale=1.0 (default)
exp_array = np.random.exponential(scale=1.0, size=(4, 4))
print("4x4 Exponential Array:\n", exp_array)
```

```python
import matplotlib.pyplot as plt

# Generate 100000x1 arrays from each distribution
exp_data = np.random.exponential(scale=1.0, size=100000)
uniform_data = np.random.rand(100000)
normal_data = np.random.randn(100000)

# Plot all distributions
plt.figure(figsize=(10, 6))

plt.hist(exp_data, density=True, bins=100, histtype="step", color="red", label="Exponential")
plt.hist(uniform_data, density=True, bins=100, histtype="step", color="blue", label="Uniform")
plt.hist(normal_data, density=True, bins=100, histtype="step", color="green", label="Normal")

plt.axis([-2.5, 8, 0, 1.2])
plt.legend(loc="upper right")
plt.title("Random Distributions Comparison")
plt.xlabel("Value")
plt.ylabel("Density")
plt.grid(True)
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D

x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = X**2 + Y**2

fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')

ax.plot_surface(X, Y, Z, cmap='viridis')
ax.set_title("3D Plot of Z = X² + Y²")
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("Z")
plt.show()
```

```python
import seaborn as sns
!wget -q https://elitedatascience.com/wp-content/uploads/2022/07/Pokemon.csv
df = pd.read_csv('Pokemon.csv', index_col=0, encoding='latin')

df.head()

stats_df = df.select_dtypes(include=['number'])
corr = stats_df.corr()
corr

plt.figure(figsize=(9,8))
sns.heatmap(corr, annot=True, fmt=".2f", cmap="coolwarm")
plt.title("Pearson Standard Coefficient and Spearman Rank Correlation")
plt.show()
```
