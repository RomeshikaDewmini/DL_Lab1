```python
import pandas as pd

df = pd.read_csv("Au_nanoparticle_dataset.csv")
df.head()
```

```python
# Select only the 4 columns
new_df = df[['N_total', 'N_bulk', 'N_surface', 'R_avg']]
```

```python
new_df.head(20)
```

```python
# Mean
mean_values = new_df.mean()

# Standard Deviation
std_values = new_df.std()

# Quartiles (25%, 50%, 75%)
quartiles = new_df.quantile([0.25, 0.5, 0.75])

# Display results
print("Mean:\n", mean_values)
print("\nStandard Deviation:\n", std_values)
print("\nQuartiles:\n", quartiles)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(20, 4))

for i, column in enumerate(new_df.columns):
    plt.subplot(1, 4, i + 1)
    plt.hist(new_df[column], bins=30, color='skyblue', edgecolor='black')
    plt.title(f'Histogram of {column}')
    plt.xlabel(column)
    plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

```python
import seaborn as sns

sns.pairplot(new_df)
```

```python
g = sns.PairGrid(new_df)

# Diagonal: Histogram + KDE
g.map_diag(sns.histplot, kde=True, color="purple")

# Lower: Bivariate KDE
g.map_lower(sns.kdeplot, cmap="Blues", fill=True)

# Upper: Regular scatterplot (optional)
g.map_upper(sns.scatterplot)

plt.show()
```
